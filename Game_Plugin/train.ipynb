```python
from ultralytics import YOLO
import cv2
import os
```

```python
# Load a pretrained YOLO model (recommended for training)
model = YOLO('yolov8n.pt')
model.to('cuda')
```

```python
os.getcwd()
```

```python
# Train the model
# os.chdir('data')
PATH = os.path.join('data','apex.yaml')
print(PATH)
results = model.train(data=PATH, epochs=30, batch=8, device=0, workers=0)
print("=============================================================")
print("====================== train =================================")
print(results)
print("=============================================================")
print("=============================================================")

# Evaluate the model's performance on the validation set
# results = model.val()
# print("=============================================================")
# print("======================= valid ================================")
# print(results)
# print("=============================================================")
# print("=============================================================")
```
